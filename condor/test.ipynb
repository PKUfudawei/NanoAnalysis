```python
import glob
import yaml
import subprocess

dataset_cards = set(glob.glob('../datasets/*/2018/*/*.yaml'))

for c in dataset_cards:
    with open(c, 'r') as f:
        d = yaml.load(f, Loader=yaml.FullLoader)
        for k,v in d.items():
            print(v)
            query = f"\"parent dataset={v} system=rucio\""
            output = subprocess.check_output(f"/cvmfs/cms.cern.ch/common/dasgoclient -query={query} -json", shell=True, encoding='utf-8')
            print(output)
```
